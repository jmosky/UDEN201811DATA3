```python
import requests
import json
```

```python
data = requests.get('https://raw.githubusercontent.com/jeffreylancaster/game-of-thrones/master/data/characters.json')
data = data.json().get('characters') # this was a key THE ONLY key
```

```python
data
```

```python
# characterName

char_to_kills = {}

for char in data:
    if "killed" in char:
        char_name = char.get('characterName')
        char_to_kills[char_name] = char.get('killed')
```

```python
for k in sorted(char_to_kills, key=lambda k: len(char_to_kills[k]), reverse=True):
    print(k, char_to_kills[k], len(char_to_kills[k]))
    print()
    break
```

```python
still_around = []


for char in data:
    # if key not in dictionary
    if "killedBy" not in char:
        char_name = char.get('characterName')
        still_around.append(char_name)
```

```python
killed_oc = 0
killed_s = 0

not_killed_oc = 0
not_killed_s = 0


for char in data:
    if "killed" in char and "siblings" in char:
        killed_s += 1
    elif "killed" in char and "siblings" not in char:
        killed_oc += 1
    elif "killed" not in char and "siblings" in char:
        not_killed_s +=1
    else:
        not_killed_oc += 1
```

```python
# group all folks by their last name into a dict:
# siblings and parents

# {last name: [charname, charname]}

from collections import defaultdict

families = defaultdict(set)

for char in data:
    relatives = []
    
    char_name = char.get('characterName')
    char_last = char_name.split()[-1]
    
    char_siblings = char.get('siblings')
    char_parents = char.get('parents')
    
    if char_siblings:
        # if char_siblings is not none
        relatives.extend(char_siblings)
    if char_parents:
        relatives.extend(char_parents)
        
    for r in relatives:
        if char_last == r.split()[-1]:
            families[char_last].add(r)
            
    families[char_last].add(char_name)

        
    # relatives should be a list of first and last names as a string
    
```
